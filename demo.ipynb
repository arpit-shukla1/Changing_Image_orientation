```python
from keras.models import load_model
import pandas as pd
import cv2
import numpy as np
from rotate_images import rotate
import random
import matplotlib.pyplot as plt
```

```python
model = load_model('model.h5')
```

```python
target = pd.read_csv('./train.truth.csv')
names = target['fn']

images = []
for name in names:
    aux = './train/' + name
    images.append(cv2.imread(aux,1))
```

```python
samples = random.sample(images,5)
```

```python
for i in samples:
    plt.figure()
    plt.imshow(i)
```

```python
for i in samples:
    img = rotate(model,i)
    plt.figure()
    plt.imshow(img)
```
